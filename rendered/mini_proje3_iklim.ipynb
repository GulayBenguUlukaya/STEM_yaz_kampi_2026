```python
import os
from python_code.helper_tr import *
import pandas as pd
import numpy as np
import plotly.express as px

iklim = veri_yukle('input_data/iklim_co2.csv')
print(f'\nÜlkeler: {iklim["ulke"].unique().tolist()}')
print(f'Yıl aralığı: {iklim["yil"].min()} - {iklim["yil"].max()}')
```

---
### 👥 Grup Tartışması — Başlamadan Önce
- Türkiye'nin CO₂ salımı son 70 yılda nasıl değişti sizce — arttı, azaldı, yoksa sabit mi? Grupça **tahmin** edin, sonra veriye bakacağız.
- 'Karbon ayak izi' ne demek? Kendi günlük hayatınızda en çok neyden kaynaklanıyor olabilir?

## Adım 1: Türkiye'nin CO₂ Tarihçesi

```python
tr = iklim[iklim['ulke'] == 'Türkiye']
fig = px.line(tr, x='yil', y='co2_milyon_ton', markers=True,
              title='Türkiye Yıllık CO₂ Emisyonu (Milyon Ton)')
fig.show()

ilk = tr[tr['yil'] == 1950]['co2_milyon_ton'].values[0]
son = tr[tr['yil'] == 2024]['co2_milyon_ton'].values[0]
print(f'1950: {ilk:.1f} milyon ton  →  2024: {son:.1f} milyon ton  →  {son/ilk:.1f}x artış!')
```

**Görev:** Hangi yıllar en hızlı artış? Hangi yıllar yavaşladı/azaldı? (1973 petrol krizi, 2001 ekonomik kriz, 2020 COVID izleri var mı?)

## 🦠 Adım 2: COVID-2020 — Doğal Bir İklim Deneyi

**Hipotez:** Pandemi dünyada tüm faaliyeti durdurdu. CO₂ ne kadar düştü?

```python
# Tüm ülkeler için 2019 → 2020 değişim
for ulke in ['Türkiye', 'Almanya', 'ABD', 'Çin', 'Hindistan', 'Dünya']:
    veri_2019 = iklim[(iklim['ulke']==ulke) & (iklim['yil']==2019)]['co2_milyon_ton'].values
    veri_2020 = iklim[(iklim['ulke']==ulke) & (iklim['yil']==2020)]['co2_milyon_ton'].values
    veri_2021 = iklim[(iklim['ulke']==ulke) & (iklim['yil']==2021)]['co2_milyon_ton'].values
    if len(veri_2019) and len(veri_2020):
        dusus = (veri_2020[0] - veri_2019[0]) / veri_2019[0] * 100
        sicrama = (veri_2021[0] - veri_2020[0]) / veri_2020[0] * 100 if len(veri_2021) else 0
        print(f'{ulke:12}: 2019→2020 {dusus:+.1f}%, 2020→2021 {sicrama:+.1f}% (geri tepki)')
```

**Görev:** Hangi ülke en çok düştü? Pandemi sonrası geri tepki ne kadar oldu? Bu "karantina iklim için yeterli mi?" sorusunun cevabıdır.

## Adım 3: Toplam vs Kişi Başı — Adalet Sorusu

```python
secili = iklim[iklim['ulke'].isin(['Türkiye', 'Almanya', 'Yunanistan', 'İran', 'ABD', 'Çin'])]
fig = px.line(secili, x='yil', y='co2_milyon_ton', color='ulke',
              title='Toplam CO₂ — Çin & ABD baskın')
fig.show()
fig = px.line(secili, x='yil', y='co2_kisi_basi', color='ulke',
              title='Kişi Başı CO₂ — ABD ve Almanya baskın, Çin orta')
fig.show()
```

```python
# 2024 son durum karşılaştırma
son_2024 = iklim[iklim['yil'] == 2024][['ulke', 'co2_milyon_ton', 'co2_kisi_basi']].sort_values('co2_milyon_ton', ascending=False)
print('2024 yılı sıralaması:')
print(son_2024.to_string(index=False))
```

**Görev:** Çin toplam #1 ama kişi başı sıralaması nerede? Bu "adaletli ölçü hangisi?" sorusunun temeli.

---
### 👥 Grup Tartışması
- Bir ülke 'toplamda az kirletiyorum' derken **kişi başına** çok kirletiyor olabilir. İklim sorumluluğu nasıl ölçülmeli — **toplam** mı, **kişi başı** mı? Grupça ortak bir görüşe varmaya çalışın.
- Kalabalık bir ülke ile küçük bir ülkeyi karşılaştırırken hangi ölçü daha adil?

## 📊 Adım 4: Pareto Analizi — Az Ülke, Çok Emisyon

```python
# 2024'te en kirletici ülkelerin payı
tum_ulkeler = iklim[(iklim['yil'] == 2024) & (iklim['ulke'] != 'Dünya')].sort_values('co2_milyon_ton', ascending=False)
tum_ulkeler['kumulatif_pay'] = tum_ulkeler['co2_milyon_ton'].cumsum() / tum_ulkeler['co2_milyon_ton'].sum() * 100
print('Pareto: Sıralı ülkelerin kümülatif yüzdesi:')
print(tum_ulkeler[['ulke', 'co2_milyon_ton', 'kumulatif_pay']].head(10).round(1).to_string(index=False))
```

**Görev:** En kirletici 2 ülke veri setimizdeki toplam %___ üretiyor. Bu adil mi? "Az ülke = çok sorumluluk" tartışması.

## 🌡️ Adım 5: CO₂ → Gerçek Sıcaklık Etkisi

OWID verisindeki `sicaklik_degisimi` sütunu, her ülkenin CO₂ emisyonunun küresel sıcaklığa katkısını gösterir (°C cinsinden).

```python
fig = px.line(secili, x='yil', y='sicaklik_degisimi', color='ulke',
              title='Her Ülkenin CO₂\'sinin Küresel Sıcaklığa Toplam Katkısı (°C)')
fig.show()

# 2024 sıralama
son = iklim[(iklim['yil'] == 2024) & (iklim['ulke'] != 'Dünya')].dropna(subset=['sicaklik_degisimi'])
son = son.sort_values('sicaklik_degisimi', ascending=False)
print('\n2024: Her ülkenin küresel ısınmaya kümülatif katkısı:')
print(son[['ulke', 'sicaklik_degisimi']].head(8).to_string(index=False))
```

**Görev:** Türkiye küresel ısınmaya kaç °C katkıda bulundu? Almanya ile karşılaştırın (sanayi tarihi farkı!)

## 🇩🇪 Adım 6: Almanya — Düşürebilen Tek Büyük Ülke

Almanya 1990'larda CO₂'sini azaltmaya başladı (yenilenebilir enerji + Doğu Almanya kapatmaları).

```python
# Almanya zirvesi ve düşüşü
alm = iklim[iklim['ulke'] == 'Almanya']
zirve = alm.loc[alm['co2_milyon_ton'].idxmax()]
print(f'🇩🇪 Almanya zirvesi: {int(zirve["yil"])} yılında {zirve["co2_milyon_ton"]:.0f} Mt')
print(f'🇩🇪 2024 değer: {alm[alm["yil"]==2024]["co2_milyon_ton"].values[0]:.0f} Mt')
print(f'🇩🇪 Zirveden düşüş: -%{(1 - alm[alm["yil"]==2024]["co2_milyon_ton"].values[0]/zirve["co2_milyon_ton"])*100:.0f}')
print(f'\n🇹🇷 Türkiye zirvesi: 2024 (henüz düşmedi)')
print(f'🇹🇷 Türkiye 2024 değer: {tr[tr["yil"]==2024]["co2_milyon_ton"].values[0]:.0f} Mt')
```

**Görev:** Almanya nasıl düşürdü, Türkiye niye hala artıyor? Yenilenebilir enerji + nükleer + verimlilik = Almanya formülü

---
### 👥 Grup Tartışması
- Almanya düşürdü, Türkiye hâlâ artıyor. Türkiye'nin emisyonu düşürmesi için **2 somut öneri** üretin.
- Bu önerilerin maliyeti kime düşer (vatandaş, devlet, sanayi)? **Adil bir geçiş** nasıl olur?

## Adım 7: 2030 Lineer Tahmin

```python
from sklearn.linear_model import LinearRegression
egitim = tr[(tr['yil'] >= 2000) & (tr['yil'] <= 2022)].dropna(subset=['co2_milyon_ton'])
model = LinearRegression().fit(egitim[['yil']], egitim['co2_milyon_ton'])
tahmin_2030 = model.predict([[2030]])[0]
tahmin_2050 = model.predict([[2050]])[0]
print(f'Türkiye 2030 lineer tahmin: {tahmin_2030:.0f} Mt')
print(f'Türkiye 2050 lineer tahmin: {tahmin_2050:.0f} Mt')
print(f'\n⚠️ Türkiye Paris Anlaşması: 2053 NET-SIFIR taahhüt etti.')
print(f'   Lineer model 2050\'de {tahmin_2050:.0f} Mt diyor — ÇELİŞKİ!')
```

**Görev:** Lineer trend ile Paris Anlaşması taahhüdü (2053 net-sıfır) ne kadar uyumsuz? Hangi yıl müdahale gerekir?
