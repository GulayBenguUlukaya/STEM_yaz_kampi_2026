## A5. Metinlerle (string) oynamak

## A8. Hazır fonksiyonlar — `len`, `sum`, `min`, `max`, `round`, `sorted`
Python'da hazır gelen, çok işe yarayan fonksiyonlar:

````python
notlar = [85, 90, 72, 60, 95, 88]

print("Kaç not  :", len(notlar))
print("Toplam   :", sum(notlar))
print("En düşük :", min(notlar))
print("En yüksek:", max(notlar))
print("Ortalama :", round(sum(notlar) / len(notlar), 1))   # round(sayı, basamak)
print("Sıralı   :", sorted(notlar))
````

## A9. Koşullar — `if` / `elif` / `else`
Bilgisayara "eğer ... ise şunu yap, değilse bunu yap" dedirtiriz. Karşılaştırma sembolleri: `==` (eşit), `!=` (eşit değil), `<`, `>`, `<=`, `>=`.

````python
not_ortalamasi = 78

if not_ortalamasi >= 85:
    print("Pekiyi 🌟")
elif not_ortalamasi >= 70:
    print("İyi 🙂")
else:
    print("Geliştirmen lazım 💪")
````

💡 `and` (ve), `or` (veya), `not` (değil) ile koşulları birleştirebilirsin: `if yas >= 13 and yas <= 19:` → genç.

🎯 **Senin sıran:** Bir sayı pozitif mi, negatif mi, sıfır mı?

````python
# ✏️ sayi'yı değiştir:
sayi = -4
if sayi > 0:
    print("Pozitif")
elif sayi < 0:
    print("Negatif")
else:
    print("Sıfır")
````

## A10. Döngüler — `for` ile tekrar tekrar
Bir işi her eleman için tekrarlamak istediğimizde **döngü** kullanırız. `range(n)` 0'dan n-1'e kadar sayar.

````python
# Bir listedeki her şehri tek tek yazdır:
sehirler = ["Malatya", "İstanbul", "İzmir"]
for sehir in sehirler:
    print("Şehir:", sehir)

print("---")

# range ile 1'den 5'e kadar:
for i in range(1, 6):
    print(i, "kere kare:", i ** 2)
````

🎯 **Senin sıran:** Bir döngüyle 1'den 10'a kadar olan sayıların toplamını bul.

````python
toplam = 0
for sayi in range(1, 11):     # 1, 2, ..., 10
    toplam = toplam + sayi    # her adımda ekle
print("1'den 10'a kadar toplam:", toplam)
````

## A11. Kendi fonksiyonunu yaz — `def`
Bir işi bir kez tanımlayıp tekrar tekrar kullanmak için **fonksiyon** yazarız. `def` ile tanımlanır, `return` ile sonuç döndürür.

````python
# selamla adında bir fonksiyon tanımlayalım:
def selamla(isim):
    return f"Merhaba {isim}, kampa hoş geldin! 🎉"

# Şimdi istediğimiz kadar kullanabiliriz:
print(selamla("Zeynep"))
print(selamla("Ahmet"))

# Sayı döndüren bir fonksiyon:
def kare_al(x):
    return x * x

print("5'in karesi:", kare_al(5))
````

🎯 **Senin sıran:** İki sayıyı toplayan bir fonksiyon yaz.

````python
# ✏️ return satırını tamamla (a + b):
def topla(a, b):
    return a + b

print("3 + 7 =", topla(3, 7))
````

## A12. Sözlükler — `dict` (etiketli bilgi)
Liste sırayla tutar; **sözlük** ise *anahtar → değer* şeklinde etiketli tutar. Süslü parantez `{}` kullanılır.

## 🏆 Mini Meydan Okuma — Öğrendiklerini birleştir!
Aşağıda bazı illerin nüfusu var. Bir **döngü** ve bir **koşul** kullanarak nüfusu 1 milyondan büyük illeri yazdıralım. (Bu, birazdan yapacağımız veri analizinin "elle" hali!)

````python
nufuslar = {"Malatya": 742725, "İstanbul": 15655924, "Bayburt": 81910, "İzmir": 4479525}

print("Nüfusu 1 milyondan büyük iller:")
for il in nufuslar:
    if nufuslar[il] > 1000000:
        print(" -", il, "→", nufuslar[il], "kişi")
````

Türkiye'nin **81 ilinin demografi verisini** (TÜİK 2023) açacağız. Her il için: bölge, nüfus, doğurganlık oranı, ortalama yaş.

💡 Bu veri, A bölümündeki sözlüğe benzer ama **81 il × 5 bilgi** içeriyor. Elle döngü yazmak yerine **pandas** denen güçlü bir araç kullanacağız.

## 2. Filtreleme Alıştırması
### 📝 Çalışma Kitapçığı — Gün 1, Bölüm 2

**Filtreleme**, bir koşulu sağlayan satırları seçmektir — tıpkı A9'daki `if` gibi, ama tüm tabloya birden uygulanır.

````python
# Bölgesi 'Doğu Anadolu' olan illeri seç.
# ✏️ İstersen başka bölge yaz: 'Marmara' · 'Ege' · 'Akdeniz' · 'İç Anadolu' · 'Karadeniz' · 'Doğu Anadolu' · 'Güneydoğu Anadolu'
dogu_anadolu = df[df['bolge'] == 'Doğu Anadolu']

print('Bu bölgede', len(dogu_anadolu), 'il var:')
print(list(dogu_anadolu['il']))
````

**📝 a)** Yukarıdaki listeden **en az 3 il** seçip Çalışma Kitapçığı'na yaz.

Şimdi **b)** şıkkı: Malatya, bu bölgenin ortalamasının üstünde mi altında mı?

````python
# Doğu Anadolu'nun ortalama nüfusunu hesapla ve Malatya ile karşılaştır.
ortalama = dogu_anadolu['nufus'].mean()
malatya  = df[df['il'] == 'Malatya']['nufus'].values[0]

print(f'Doğu Anadolu ortalama nüfusu: {ortalama:.0f} kişi')
print(f'Malatya nüfusu             : {malatya} kişi')

if malatya > ortalama:
    print('➡️  Malatya, ortalamanın ÜSTÜNDE')
else:
    print('➡️  Malatya, ortalamanın ALTINDA')
````

💡 **İsteğe bağlı:** Yukarıda `'nufus'` yerine `'dogurganlik_orani'` veya `'ortalama_yas'` yazıp tekrar çalıştırırsan o ölçüde karşılaştırırsın.

🧑‍🏫 **Cevap:** Malatya nüfusta ortalamanın **üstünde**.

## 🆘 Bir hata mı aldın? Sık karşılaşılanlar
- **`SyntaxError`** → Bir tırnak, parantez veya iki nokta `:` eksik olabilir. Satırı dikkatle kontrol et.
- **`NameError: name '...' is not defined`** → Değişkeni yazmadan kullandın ya da yukarıdaki hücreyi çalıştırmadın.
- **`FileNotFoundError`** → Drive bağlandı mı? **0. Hazırlık** hücresini çalıştırdın mı?
- **`KeyError: 'Bolge'`** → Sütun adları küçük harf duyarlı; doğrusu `'bolge'`. `df.columns` ile bakabilirsin.
- **Türkçe karakter** → Filtrelerken `'Doğu Anadolu'` yazımını birebir kopyala.
